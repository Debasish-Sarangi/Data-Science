```python
def main():
    import numpy as np
    from itertools import product
    number_of_people = int(input().strip())
    Red_Orange= int(input().strip())
    Red_Green= int(input().strip())
    Orange_Red= int(input().strip())
    Orange_Green = int(input().strip())
    Green_Red= int(input().strip())
    Green_Orange= int(input().strip())
    counter=[]
    cnt=0
    red=0
    mn=0
    rnd=round(number_of_people/2)
    if(Red_Orange!=0):

        mn=min(rnd,Red_Orange)
        rand_people=np.repeat(['R','O'],mn)
        red=red+((mn**2))

        a=product(rand_people,repeat=2)

        for h in a:
            if(not(h[0]=='O' and h[1]=='R')):
                cnt=cnt+1


    if(Red_Green!=0):
        mn=min(rnd,Red_Green)
        rand_people=np.repeat(['R','G'],mn)
        red=red+((mn**2))
        a=product(rand_people,repeat=2)

        for h in a:
            if(not(h[0]=='G' and h[1]=='R') ):
                counter.append(h)


    if(Orange_Red!=0):
        mn=min(rnd,Orange_Red)
        rand_people=np.repeat(['O','R'],mn)
        red=red+((mn**2))
        a=product(rand_people,repeat=2)

        for h in a:
            if(not(h[0]=='R' and h[1]=='O')):
                counter.append(h)

    if(Orange_Green!=0):
        mn=min(rnd,Orange_Green)
        rand_people=np.repeat(['O','G'],mn)
        red=red+((mn**2))
        a=product(rand_people,repeat=2)

        for h in a:
            if(not(h[0]=='G' and h[1]=='O')):
                counter.append(h)

    if(Green_Red!=0):
        mn=min(rnd,Green_Red)
        rand_people=np.repeat(['G','R'],mn)
        red=red+((mn**2))
        a=product(rand_people,repeat=2)

        for h in a:
            if(not(h[0]=='R' and h[1]=='G')):
                counter.append(h)

    if(Green_Orange!=0):
        mn=min(rnd,Green_Orange)
        rand_people=np.repeat(['G','O'],mn)
        red=red+((mn**2))
        a=product(rand_people,repeat=2)

        for h in a:
            if(not(h[0]=='O' and h[1]=='G')):
                counter.append(h)

    print((len(counter)-red)+2)
main()
```
